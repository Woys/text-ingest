```python
import asyncio
import csv
import json
import re
import sys
from collections import Counter, defaultdict
from datetime import date, timedelta
from pathlib import Path


def _find_project_src(start: Path) -> Path:
    for base in [start, *start.parents]:
        candidate = base / "src" / "data_ingestion"
        if candidate.exists():
            return base / "src"

    two_up = start.parent.parent
    candidate = two_up / "src" / "data_ingestion"
    if candidate.exists():
        return two_up / "src"

    raise RuntimeError("Could not locate project src/data_ingestion from notebook cwd")


SRC_PATH = _find_project_src(Path.cwd().resolve())
src_path_str = str(SRC_PATH)
if src_path_str in sys.path:
    sys.path.remove(src_path_str)
sys.path.insert(0, src_path_str)

for module_name in list(sys.modules):
    if module_name == "data_ingestion" or module_name.startswith("data_ingestion."):
        del sys.modules[module_name]

import data_ingestion

print(f"Using data_ingestion from {Path(data_ingestion.__file__).resolve()}")

import pandas as pd

from data_ingestion.exceptions import QuotaExceededError
from data_ingestion.pipeline import async_stream_transformed_records
```

```python
STOPWORDS = {
    "about",
    "after",
    "again",
    "against",
    "all",
    "also",
    "and",
    "are",
    "because",
    "been",
    "being",
    "between",
    "but",
    "can",
    "could",
    "from",
    "has",
    "have",
    "into",
    "its",
    "more",
    "new",
    "not",
    "over",
    "than",
    "that",
    "the",
    "their",
    "then",
    "there",
    "these",
    "this",
    "through",
    "under",
    "using",
    "was",
    "were",
    "when",
    "where",
    "which",
    "while",
    "with",
    "would",
    "your",
}

LOW_MEMORY_COLUMNS = [
    "source",
    "external_id",
    "title",
    "authors",
    "published_date",
    "url",
    "abstract",
    "full_text_url",
    "topic",
    "record_type",
    "publication",
    "summary_text",
]


def first_text_for_export(*values):
    for value in values:
        if isinstance(value, str) and value.strip():
            return value.strip()
        if isinstance(value, list):
            for item in value:
                if isinstance(item, str) and item.strip():
                    return item.strip()
    return None


def openalex_publication_from_raw(raw):
    for location_key in ("primary_location", "best_oa_location"):
        location = raw.get(location_key) or {}
        source = location.get("source") or {}
        name = first_text_for_export(
            source.get("display_name"), source.get("display_name_alternatives")
        )
        if name:
            return name
    host_venue = raw.get("host_venue") or {}
    return first_text_for_export(host_venue.get("display_name"))


def publication_for_record(record):
    raw = record.raw_payload if isinstance(record.raw_payload, dict) else {}
    source = record.source

    if source == "crossref":
        return (
            first_text_for_export(raw.get("container-title"), raw.get("publisher"))
            or "Crossref"
        )
    if source == "openalex":
        return openalex_publication_from_raw(raw) or "OpenAlex"
    if source == "googlenews":
        title = record.title or ""
        if " - " in title:
            return title.rsplit(" - ", 1)[-1].strip() or "Google News"
        return "Google News"
    if source == "edgar":
        src = raw.get("_source") or {}
        return first_text_for_export(src.get("display_names")) or "SEC EDGAR"
    if source == "federalregister":
        agencies = raw.get("agencies") or []
        for agency in agencies:
            name = first_text_for_export(agency.get("raw_name"), agency.get("name"))
            if name:
                return name
        return "Federal Register"
    if source == "stackexchange":
        return STACKEXCHANGE_SITE
    if source == "hackernews":
        return "Hacker News"
    return source or "unknown"


def record_summary_text(record):
    parts = [record.title, record.abstract]
    return " ".join(
        part.strip() for part in parts if isinstance(part, str) and part.strip()
    )


source_counts = Counter()
publication_counts = Counter()
topic_counts_counter = Counter()
term_counts = Counter()
publication_texts = defaultdict(list)
publication_titles = defaultdict(list)
publication_sources = defaultdict(set)
run_notes = []
write_lock = asyncio.Lock()


def update_low_memory_stats(row):
    source_counts[row["source"]] += 1
    publication_counts[row["publication"]] += 1
    topic_counts_counter[row["topic"] or "unknown"] += 1
    for token in re.findall(r"[a-z][a-z0-9_-]{2,}", row["summary_text"].lower()):
        if token not in STOPWORDS:
            term_counts[token] += 1

    publication = row["publication"]
    if row["summary_text"] and len(publication_texts[publication]) < 25:
        publication_texts[publication].append(row["summary_text"])
    if row["title"] and len(publication_titles[publication]) < 5:
        publication_titles[publication].append(row["title"])


async def fetch_source(spec, writer):
    source_name = spec["source"]
    records_kept = 0
    try:
        async for source, record in async_stream_transformed_records(
            [spec],
            transform_spec=transform_spec,
            start_date=YESTERDAY_ISO,
            end_date=YESTERDAY_ISO,
            concurrent_sources=False,
            max_async_queue_size=MAX_ASYNC_QUEUE_SIZE,
        ):
            row = {
                "source": source,
                "external_id": record.external_id,
                "title": record.title,
                "authors": json.dumps(record.authors, ensure_ascii=False),
                "published_date": (
                    record.published_date.isoformat()
                    if record.published_date is not None
                    else None
                ),
                "url": record.url,
                "abstract": record.abstract,
                "full_text_url": record.full_text_url,
                "topic": record.topic,
                "record_type": record.record_type.value,
                "publication": publication_for_record(record),
                "summary_text": record_summary_text(record),
            }
            async with write_lock:
                writer.writerow(row)
                update_low_memory_stats(row)
            records_kept += 1
    except QuotaExceededError as exc:
        return {
            "source": source_name,
            "status": "quota_exhausted",
            "records_kept": records_kept,
            "message": str(exc),
        }
    except Exception as exc:
        return {
            "source": source_name,
            "status": "failed",
            "records_kept": records_kept,
            "message": str(exc),
        }

    return {
        "source": source_name,
        "status": "completed",
        "records_kept": records_kept,
        "message": "",
    }


with RECORDS_OUTPUT.open("w", encoding="utf-8", newline="") as fh:
    writer = csv.DictWriter(fh, fieldnames=LOW_MEMORY_COLUMNS)
    writer.writeheader()
    semaphore = asyncio.Semaphore(MAX_SOURCE_CONCURRENCY)

    async def fetch_source_bounded(spec):
        async with semaphore:
            return await fetch_source(spec, writer)

    run_notes = await asyncio.gather(
        *(fetch_source_bounded(spec) for spec in fetcher_specs)
    )

run_notes_df = pd.DataFrame(run_notes)
print(f"Wrote low-memory export to {RECORDS_OUTPUT}")
display(run_notes_df)

if LOAD_EXPORTED_FOR_ANALYSIS:
    df = pd.read_csv(RECORDS_OUTPUT)
else:
    df = pd.DataFrame(columns=LOW_MEMORY_COLUMNS)
    print(
        "Set LOAD_EXPORTED_FOR_ANALYSIS = True and rerun this cell "
        "to load all rows into pandas."
    )

low_memory_records_by_source = pd.DataFrame(
    source_counts.most_common(), columns=["source", "records"]
)
low_memory_top_publications = pd.DataFrame(
    publication_counts.most_common(25), columns=["publication", "records"]
)
low_memory_topic_counts = pd.DataFrame(
    topic_counts_counter.most_common(25), columns=["topic", "records"]
)
low_memory_top_terms = pd.DataFrame(
    term_counts.most_common(25), columns=["term", "count"]
)
run_notes_df
```

```python
STOPWORDS = {
    "about",
    "after",
    "again",
    "against",
    "all",
    "also",
    "and",
    "are",
    "because",
    "been",
    "being",
    "between",
    "but",
    "can",
    "could",
    "from",
    "has",
    "have",
    "into",
    "its",
    "more",
    "new",
    "not",
    "over",
    "than",
    "that",
    "the",
    "their",
    "then",
    "there",
    "these",
    "this",
    "through",
    "under",
    "using",
    "was",
    "were",
    "when",
    "where",
    "which",
    "while",
    "with",
    "would",
    "your",
}


def first_text(*values):
    for value in values:
        if isinstance(value, str) and value.strip():
            return value.strip()
        if isinstance(value, list):
            for item in value:
                if isinstance(item, str) and item.strip():
                    return item.strip()
    return None


def openalex_publication(raw):
    for location_key in ("primary_location", "best_oa_location"):
        location = raw.get(location_key) or {}
        source = location.get("source") or {}
        name = first_text(
            source.get("display_name"), source.get("display_name_alternatives")
        )
        if name:
            return name
    host_venue = raw.get("host_venue") or {}
    return first_text(host_venue.get("display_name"))


def publication_for_row(row):
    raw = row.get("raw_payload") if isinstance(row.get("raw_payload"), dict) else {}
    source = row.get("source")

    if source == "crossref":
        return (
            first_text(raw.get("container-title"), raw.get("publisher")) or "Crossref"
        )
    if source == "openalex":
        return openalex_publication(raw) or "OpenAlex"
    if source == "googlenews":
        title = row.get("title") or ""
        if " - " in title:
            return title.rsplit(" - ", 1)[-1].strip() or "Google News"
        return "Google News"
    if source == "edgar":
        src = raw.get("_source") or {}
        return first_text(src.get("display_names")) or "SEC EDGAR"
    if source == "federalregister":
        agencies = raw.get("agencies") or []
        for agency in agencies:
            name = first_text(agency.get("raw_name"), agency.get("name"))
            if name:
                return name
        return "Federal Register"
    if source == "stackexchange":
        return STACKEXCHANGE_SITE
    if source == "hackernews":
        return "Hacker News"
    return source or "unknown"


def make_summary_text(row):
    parts = [row.get("title"), row.get("abstract"), row.get("full_text")]
    return " ".join(
        str(part).strip() for part in parts if isinstance(part, str) and part.strip()
    )


def tokenize(text):
    for token in re.findall(r"[a-z][a-z0-9_-]{2,}", str(text).lower()):
        if token not in STOPWORDS:
            yield token


def extractive_summary(texts, max_sentences=3):
    joined = " ".join(
        str(text) for text in texts if isinstance(text, str) and text.strip()
    )
    sentences = [
        s.strip() for s in re.split(r"(?<=[.!?])\\s+|\\n+", joined) if s.strip()
    ]
    if not sentences:
        return ""

    term_counts = Counter(tokenize(joined))
    scored = []
    for idx, sentence in enumerate(sentences):
        terms = list(tokenize(sentence))
        if not terms:
            continue
        score = sum(term_counts[t] for t in terms) / max(len(terms), 1)
        scored.append((score, idx, sentence))

    selected = sorted(scored, reverse=True)[:max_sentences]
    selected = sorted(selected, key=lambda item: item[1])
    return " ".join(sentence for _, _, sentence in selected)


if not df.empty:
    df["publication"] = df.apply(publication_for_row, axis=1)
    df["summary_text"] = df.apply(make_summary_text, axis=1)
else:
    df["publication"] = pd.Series(dtype="object")
    df["summary_text"] = pd.Series(dtype="object")

df[["source", "publication", "title", "published_date", "url"]].head(10)
```

```python
top_publications = (
    df["publication"]
    .fillna("unknown")
    .value_counts()
    .head(25)
    .rename_axis("publication")
    .reset_index(name="records")
    if not df.empty
    else low_memory_top_publications
)
top_publications
```

```python
if not df.empty:
    topic_counts = (
        df["topic"]
        .fillna("unknown")
        .value_counts()
        .head(25)
        .rename_axis("topic")
        .reset_index(name="records")
    )

    term_counts = Counter()
    for text in df["summary_text"].dropna() if "summary_text" in df else []:
        term_counts.update(tokenize(text))

    top_terms = pd.DataFrame(term_counts.most_common(25), columns=["term", "count"])
else:
    topic_counts = low_memory_topic_counts
    top_terms = low_memory_top_terms

display(topic_counts)
top_terms
```

```python
summary_rows = []
for publication in top_publications["publication"].head(15):
    if not df.empty:
        group = df[df["publication"] == publication].copy()
        titles = [title for title in group["title"].dropna().head(5).tolist()]
        sources = ", ".join(sorted(group["source"].dropna().unique()))
        texts = group["summary_text"].tolist()
        records = len(group)
    else:
        titles = publication_titles.get(publication, [])
        sources = ", ".join(sorted(publication_sources.get(publication, [])))
        texts = publication_texts.get(publication, [])
        records = int(
            top_publications.loc[
                top_publications["publication"] == publication, "records"
            ].iloc[0]
        )

    summary_rows.append(
        {
            "publication": publication,
            "records": records,
            "sources": sources,
            "representative_titles": " | ".join(titles),
            "extractive_summary": extractive_summary(texts),
        }
    )

publication_summaries = pd.DataFrame(
    summary_rows,
    columns=[
        "publication",
        "records",
        "sources",
        "representative_titles",
        "extractive_summary",
    ],
)
publication_summaries
```
